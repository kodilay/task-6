```python
import pandas as pd
```

```python
df = pd.read_csv("HousingData.csv")
print(df.head())
```

```python
print(df.isnull().sum())  # Hangi sütunda ne kadar eksik var 
```

```python
print(df.info())
```

```python
df = df.fillna({
    "CRIM": df["CRIM"].median(),
    "ZN": df["ZN"].median(),
    "INDUS": df["INDUS"].median(),
    "AGE": df["AGE"].median(),
    "LSTAT": df["LSTAT"].median(),
    "CHAS": df["CHAS"].mode()[0]  # En çok tekrar etmiş olan değer
})
```

```python
print(df.isnull().sum()) # Eksik olanları kontrol etme
```

```python
df.to_csv("boston_housing_cleaned.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bina yaşı ve Ev fiyatları arasındaki ilişkiyi gösteren Grafik
plt.figure(figsize=(10,5))
sns.scatterplot(x=df["AGE"], y=df["MEDV"],alpha=0.5)
    plt.xlabel("Bina Yaşı (AGE) " )
    plt.ylabel("EV Fiyatı (MEDV)" )
    plt.title("Bina yaşı ve Ev fiyatları arasındaki ilişki")
plt.show()
```

```python
# Oda fiyatlarındaki ve Ev fiyatlarındaki ortalama ilişki
plt.figure(figsize=(10,5))
sns.scatterplot(x=df["RM"], y=df["MEDV"],alpha=0.5)
plt.xlabel(" Oda Sayısı (RM) ")
plt.ylabel(" Ev Fiyatı (MEDV) ")
plt.title("Oda fiyatlarındaki ve Ev fiyatlarındaki ortalama ilişki")
plt.show()
```
